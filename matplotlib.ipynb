```python
#Draw a line in a diagram from position (0,0) to position (6,250):
#numpy is python library which is used for working with arrays
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 10])
ypoints = np.array([0, 350])

#plotting x axis and yaxis points have to use plot function which used to draw points (markers) in a diagram.
#plot function accept two  parameters one for x and y axis.
plt.plot(xpoints, ypoints)
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
#students with grade
ypoints = np.array(["ram","ravi","rex"])
xpoints = np.array ([100,75,82])
plt.xlabel("Marks")
plt.ylabel("students")
plt.plot(xpoints,ypoints)
plt.show()
```
